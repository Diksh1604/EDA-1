```python
#Ans 12

print("Dataset Overview:")
print(df.head())

if 'Date_of_Sale' not in df.columns or 'Sales' not in df.columns:
    raise ValueError("The dataset must contain 'Date_of_Sale' and 'Sales' columns.")

df['Date_of_Sale'] = pd.to_datetime(df['Date_of_Sale'], errors='coerce')

df = df.dropna(subset=['Date_of_Sale'])

df['Month'] = df['Date_of_Sale'].dt.month
df['Season'] = df['Month'].apply(lambda x: 'Winter' if x in [12, 1, 2]
                                  else 'Spring' if x in [3, 4, 5]
                                  else 'Summer' if x in [6, 7, 8]
                                  else 'Autumn')

monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar', color='lightblue')
plt.title('Monthly Car Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()

seasonal_sales = df.groupby('Season')['Sales'].sum()

plt.figure(figsize=(8, 5))
seasonal_sales.plot(kind='bar', color='orange')
plt.title('Seasonal Car Sales')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

print("\nMonthly Sales Trends:")
print(monthly_sales)

print("\nSeasonal Sales Trends:")
print(seasonal_sales)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/path/to/amazon.csv'  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
# Ans1

data['main_category'] = data['category'].str.split('|').str[0]  # Extract main category
avg_rating_per_category = data.groupby('main_category')['rating'].mean().reset_index()
print("Average Rating by Category:")
print(avg_rating_per_category)
```

```python
# Ans2

top_products_by_category = data.loc[data.groupby('main_category')['rating_count'].idxmax()]
print("\nTop Rating Count Products by Category:")
print(top_products_by_category[['main_category', 'product_name', 'rating_count']])
```

```python
# Ans3

plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='actual_price', y='discounted_price', hue='main_category')
plt.title('Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

avg_discount_by_category = data.groupby('main_category')['discount_percentage'].mean().reset_index()
print("\nAverage Discount Percentage by Category:")
print(avg_discount_by_category)
```

```python
#Ans 4

most_popular_products = data['product_name'].value_counts().head(5)
print("\nMost Popular Product Names:")
print(most_popular_products)
```

```python
#Ans 5

from collections import Counter
keywords = Counter(" ".join(data['product_name']).split())
most_common_keywords = keywords.most_common(10)
print("\nMost Popular Product Keywords:")
print(most_common_keywords)
```

```python
#Ans 6

most_popular_reviews = data['review_content'].value_counts().head(5)
print("\nMost Popular Product Reviews:")
print(most_popular_reviews)
```

```python
#Ans 7

correlation = data[['discounted_price', 'rating']].corr().iloc[0, 1]
print("\nCorrelation Between Discounted Price and Rating:", correlation)
```

```python
#Ans 8

top_categories_by_rating = avg_rating_per_category.sort_values(by='rating', ascending=False).head(5)
print("\nTop 5 Categories by Highest Ratings:")
print(top_categories_by_rating)
```

```python
#Ans 9

low_rated_products = data[data['rating'] < 3]
low_rated_categories = low_rated_products['main_category'].value_counts()
print("\nPotential Areas for Improvement (Low Rated Categories):")
print(low_rated_categories)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Ans 1

file_path = '/path/to/spotify.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Check for null values and handle them
print("Null Values in Dataset:")
print(data.isnull().sum())
data = data.dropna()  # Drop rows with null values (adjust as needed)
```

```python
#Ans 2

plt.figure(figsize=(10, 6))
sns.histplot(data['popularity'], kde=True, bins=20, color='blue')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
#Ans 3

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='duration_ms', y='popularity', color='green')
plt.title('Popularity vs. Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
#Ans 4

plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='artist_name', order=data['artist_name'].value_counts().index[:10])
plt.title('Number of Tracks by Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist Name')
plt.show()
```

```python
#Ans 5

least_popular_tracks = data.nsmallest(5, 'popularity')[['artist_name', 'track_name', 'popularity']]
print("\nTop 5 Least Popular Tracks:")
print(least_popular_tracks)
```

```python
#Ans 6

top_artists = data.groupby('artist_name')['popularity'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 Most Popular Artists by Average Popularity:")
print(top_artists)
```

```python
#Ans 7

most_popular_tracks = {}
for artist in top_artists.index:
    track = data[data['artist_name'] == artist].nlargest(1, 'popularity')[['track_name', 'popularity']]
    most_popular_tracks[artist] = track
print("\nMost Popular Tracks for Top 5 Artists:")
for artist, track in most_popular_tracks.items():
    print(f"{artist}: {track.iloc[0]['track_name']} (Popularity: {track.iloc[0]['popularity']})")
```

```python
#Ans 8

sns.pairplot(data[['popularity', 'duration_ms', 'danceability', 'energy', 'loudness']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
#Ans 9

plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='artist_name', y='duration_ms', order=data['artist_name'].value_counts().index[:10])
plt.title('Duration of Tracks Across Artists')
plt.xlabel('Artist Name')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Ans 10

plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='artist_name', y='popularity', order=data['artist_name'].value_counts().index[:10])
plt.title('Distribution of Track Popularity Across Artists')
plt.xlabel('Artist Name')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.show()
```
