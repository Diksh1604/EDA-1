```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = '/content/BIKE DETAILS.csv'
bike_data = pd.read_csv(file_path)
```

```python
#Ans 1

selling_price_range = (bike_data['selling_price'].min(), bike_data['selling_price'].max())
```

```python
#Ans 2

median_selling_price = bike_data['selling_price'].median()
```

```python
#Ans 3

most_common_seller_type = bike_data['seller_type'].mode()[0]
```

```python
#Ans 4

bikes_above_50k_km = (bike_data['km_driven'] > 50000).sum()
```

```python
#Ans 5

Average_km_by_owner = bike_data.groupby('owner')['km_driven'].mean()
```

```python
#Ans 6

proportion_2015_or_older = (bike_data['year'] <= 2015).mean()
```

```python
#Ans 7

missing_values = bike_data.isnull().sum()
```

```python
#Ans 8

highest_ex_showroom_price = bike_data['ex_showroom_price'].max()
bike_with_highest_price = bike_data[bike_data['ex_showroom_price'] == highest_ex_showroom_price]['name'].iloc[0]
```

```python
#Ans 9

bikes_by_seller_type = bike_data['seller_type'].value_counts()
```

```python
#Ans 10

first_owner_bikes = bike_data[bike_data['owner'] == '1st owner']
correlation_first_owner = first_owner_bikes[['selling_price', 'km_driven']].corr()
```

```python
#Ans 11

Q1 = bike_data['km_driven'].quantile(0.25)
Q3 = bike_data['km_driven'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

bike_data_no_outliers = bike_data[(bike_data['km_driven'] >= lower_bound) & (bike_data['km_driven'] <= upper_bound)]
```

```python
#Ans 12

plt.figure(figsize=(10, 6))
sns.scatterplot(data=bike_data_no_outliers, x='year', y='selling_price')
plt.title("Relationship between Year and Selling Price")
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.show()
```

```python
#Ans 13

current_year = 2024
bike_data['age'] = current_year - bike_data['year']
avg_depreciation = bike_data.groupby('age')['selling_price'].mean()
```

```python
#Ans 14

avg_price_by_year = bike_data.groupby('year')['selling_price'].mean()
bike_data['avg_price_by_year'] = bike_data['year'].map(avg_price_by_year)
significantly_priced_bikes = bike_data[bike_data['selling_price'] > 1.5 * bike_data['avg_price_by_year']]
```

```python
#Ans 15

correlation_matrix = bike_data_no_outliers.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
